```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

def rollavg_pandas(a,n):
        'Pandas rolling average over data set a with window size n. Returns a centered np array of same size'
        return np.ravel(pd.DataFrame(a).rolling(n, center=True, min_periods=1).mean().to_numpy())
```

```python
run_number = 5

file = r'C:\Users\Chantal\Documents\Deep-QLearning-Agent-for-Traffic-Signal-Control\resultsRun2\result5\plot_reward_data.txt'

data = np.loadtxt(file)

dynamic = data[::2]  #dynamic
low = data[1::6]
medium = data[3::6]
high = data[5::6]


#apply roling window the same length as the amount of trained on scenarios
roling_window = 6
all_averaged = rollavg_pandas(data, roling_window)


episodes = len(data)

min_val = min(data)
max_val = max(data)
```

```python
# ALL TOGETHER BUT SPLIT BY SCENARIO

plt.style.use('ggplot')
plt.rcParams.update({'font.size': 14})  # set bigger font size

fig, ax = plt.subplots()
plt.title("All scenarios")

# plt.plot(np.arange(0,episodes), all_averaged,  'ok', label = "All traffic scenarios with rolling average")
# plt.plot(np.arange(5,episodes,6), high, 'or', label = "High constant traffic scenarios")
# plt.plot(np.arange(3,episodes,6), medium, 'og', label = "Medium constant traffic scenarios")
# plt.plot(np.arange(1,episodes,6), low, 'ob', label = "Low constant traffic scenarios")
# plt.plot(np.arange(0,episodes,2), dynamic,  'oc', label = "Dynamic traffic scenarios")

# plt.plot(np.arange(0,episodes), all_averaged,  'k', label = "All traffic scenarios with rolling average")
plt.plot(np.arange(5,episodes,6), high, 'green', label = "High constant traffic scenarios")
plt.plot(np.arange(3,episodes,6), medium, 'red', label = "Medium constant traffic scenarios")
plt.plot(np.arange(1,episodes,6), low, 'orange', label = "Low constant traffic scenarios")
plt.plot(np.arange(0,episodes,2), dynamic,  'blue', label = "Dynamic traffic scenarios")

legend = ax.legend(loc='best', shadow=True, fontsize='medium')

plt.xlabel('Episode')
plt.ylabel('Cumulative negative reward')

# ax.set_facecolor('black')

plt.margins(0)
plt.ylim(min_val - 0.05 * abs(min_val), max_val + 0.05 * abs(max_val))
fig = plt.gcf()
fig.set_size_inches(20, 11.25)
plt.show()
fig.savefig(os.path.join(r'C:\Users\Chantal\Documents\Deep-QLearning-Agent-for-Traffic-Signal-Control\resultsRun2', 'plot'+str(run_number)+'.png'))
```
